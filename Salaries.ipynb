```python
df.age = df.age.astype('category')
df.timestamp = pd.to_datetime(df.timestamp)
df.industry = df.industry.astype(str)
df.job_title = df.job_title.astype(str)
df.job_title = df.job_title.str.lower()
df.industry = df.industry.str.lower()
```

```python
def flocast(x):
    try:
        float(x)
        return True
    except:
        return False
```

```python
badsal_list = list()
unus_sal = list()
idx_list = list()
for idx, x in enumerate(df.salary):
    if not flocast(x):
        badsal_list.append(x)
    elif float(x) < 1e2:
        unus_sal.append(x)
        idx_list.append(idx)
        print('idx: ',idx)
       # print(x)
print(len(badsal_list))
print(len(unus_sal))
```
